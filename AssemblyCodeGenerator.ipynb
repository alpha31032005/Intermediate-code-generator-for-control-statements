```python
def generate_assembly(tokens):
    assembly_code = []
    i = 0

    while i < len(tokens):
        token = tokens[i]

        if token[0] == 'if':
            assembly_code.append("START_IF")
            # Process condition
            i += 1  # Skip '('
            assembly_code.append(f"LOAD {tokens[i+1][0]}")  # 'a'
            assembly_code.append("COMP <")
            assembly_code.append(f"LOAD {tokens[i+3][0]}")  # 'b'
            assembly_code.append("JUMP_IF_FALSE ELSE_BLOCK")
            assembly_code.append("THEN_BLOCK:")
            i += 4  # Skip condition

            # Process then block (c = c + 1)
            assembly_code.append(f"LOAD {tokens[i][0]}")  # 'c'
            assembly_code.append("ADD 1")
            assembly_code.append(f"STORE {tokens[i][0]}")  # 'c'
            assembly_code.append("JUMP END_IF")
            assembly_code.append("ELSE_BLOCK:")
            i += 4  # Skip '} else {'

            # Process else block (exit)
            assembly_code.append("EXIT")
            assembly_code.append("END_IF:")

        i += 1  # Move to the next token

    return "\n".join(assembly_code)

# Demo input token stream
tokens = [
    ('if', 'KEYWORD'),
    ('(', 'IDENTIFIER'),
    ('a', 'IDENTIFIER'),
    ('<', 'IDENTIFIER'),
    ('b', 'IDENTIFIER'),
    (')', 'IDENTIFIER'),
    ('{', 'IDENTIFIER'),
    ('c', 'IDENTIFIER'),
    ('=', 'IDENTIFIER'),
    ('c', 'IDENTIFIER'),
    ('+', 'IDENTIFIER'),
    ('1', 'NUMBER'),
    (';', 'IDENTIFIER'),
    ('}', 'IDENTIFIER'),
    ('else', 'KEYWORD'),
    ('{', 'IDENTIFIER'),
    ('exit', 'IDENTIFIER'),
    (';', 'IDENTIFIER'),
    ('}', 'IDENTIFIER')
]

# Generate assembly code
assembly_code = generate_assembly(tokens)

# Display the generated assembly code
print("Generated Assembly Code:")
print(assembly_code)
```
